```python
import pandas as pd
import numpy as np
from sklearn.decomposition import TruncatedSVD
```

```python
reaction_dataset = pd.DataFrame(columns=['page_id', 'post_id', 'comment_body', 'user_id'])
page_names = ['bangkokeatbig', 'bornmadeak', 'DakNaeNon', 'dinewithpigs', 'paidonnnn', 'snapbeforeeatofficial', 'ipedinth', 'AroiTongShare', 'Dakmaipenyaadak', 'semagudak', 'daneating', 'badtaste.donteat', 'raosaidak', 'onnbaby', 'gin100', 'kintampro', 'kinlagthailand', 'Enjoyeatingbyme', 'aroiyummy', 'Unlucky2BDelicious', 'friendguha']
for page_name in page_names:
    temp = pd.read_csv('./data/comments/'+page_name+'.csv')
    reaction_dataset = pd.concat([reaction_dataset, temp], ignore_index=True)
reaction_dataset = reaction_dataset.iloc[:,1:]
```

```python
reaction_dataset.describe()
```

```python
for i in reaction_dataset.index:
    if i%1000==0:
        print(i)
    reaction_dataset.loc[i, 'page_id'] = reaction_dataset.loc[i, 'post_id'].split('_')[0]
reaction_dataset
```

```python
reaction_dataset['comment_body']=1
reaction_dataset
```

```python
um = pd.pivot_table(data=reaction_dataset, values='comment_body', aggfunc='sum', index='user_id', columns='page_id', fill_value=0)
```

```python
um.T
```

```python
um.sum(axis=1)
```

```python
um = um[um.sum(axis=1)>1]
# for i in um.index:
#     result = um.loc[i].sum()
#     if result==1:
#         um.drop(i, axis=0, inplace=True)
um.shape
```

```python
X = um.T

SVD = TruncatedSVD(n_components=12, random_state=17)
resultant_matrix = SVD.fit_transform(X)
resultant_matrix.shape
```

```python
corr_mat = pd.DataFrame(resultant_matrix.T).corr()
corr_mat.shape
corr_mat.values
```

```python
corr_mat = np.corrcoef(resultant_matrix)
page_ids = um.columns
for i in range(21):
    corr_mat_selected = corr_mat[i]
    print(list(page_ids)[i])
    print(list(page_ids[(corr_mat_selected<0.99) & (corr_mat_selected>0.6)]))
```

```python
from sklearn.externals import joblib
joblib.dump(corr_mat, './data/recommendation_model.pk')
joblib.dump(page_ids, './data/page_ids.pk')
```
